```python
import pandas
data = pandas.read_csv("datasets/data.csv")
data
```

```python
#check if data is empty
data.isnull().sum()
```

```python
#since the data is clean we need to split the data
#We are going to split into two
# we are going to have the X variable and Y variable
array=data.values
array
```

```python
X = array[:,0:8]
X
```

```python
#Variable Y
Y = array[:,8]

Y
```

```python
# We need to split data into training set - 70% and 30% - testing set
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
#choosing a model
#we have different types of models e.g KNN,logistic regression,decision tre,random forest,naive bayes,support vector machine.
#we use Gradient boosting classifier
from sklearn.ensemble import GradientBoostingClassifier 
model = GradientBoostingClassifier()
model.fit(X_train,Y_train)
```

```python
#We tset the model for accuracy
predictions = model.predict(X_test)
print("Model predicted", predictions)
print("Expected",Y_test)
```

```python
#we check accuracy in terms of percentage.
from sklearn.metrics import accuracy_score
print("Accuracy",accuracy_score(Y_test,predictions))
```

```python
#we are going to use random forest classifier
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()

model1.fit(X_train,Y_train)
```

```python
#Test the model for accuracy
predictions = model1.predict(X_test)
print("Model predicted", predictions)
print("Expected",Y_test)
```

```python
#we check the accuracy in terms of percentage
from sklearn.metrics import accuracy_score
print("Accuracy",accuracy_score(Y_test,predictions))
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(X_train, Y_train)
```

```python
#test model for accuracy
predictions = model2.predict(X_test)
print("Model predicted", predictions)
print("Expected",Y_test)
```

```python
#check for accuracy in percentage
from sklearn.metrics import accuracy_score
print("Accuracy", accuracy_score(Y_test, predictions))
```

```python
#Logistic REgression
from sklearn.linear_model import LogisticRegression
model3 =  LogisticRegression()
model3.fit(X_train,Y_train)
```

```python
#test the model for accuracy
predictions = model3.predict(X_test)
print("Model predicted", predictions)
print("Expected", Y_test)
```

```python
#accuracy in percentage
from sklearn.metrics import accuracy_score
print("Accuracy",accuracy_score(Y_test, predictions))
```

```python
#Support Vector Machine
from sklearn.svm import SVC
model4 = SVC()
model4.fit(X_train, Y_train)
```

```python
#test the model for accuracy
predictions = model4.predict(X_test)
print("Model predicted", predictions)
print("Expected",Y_test)
```

```python
#Accuracy in percentage
from sklearn.metrics import accuracy_score
print("Accuracy", accuracy_score(Y_test, predictions))
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
model5 = GaussianNB()
model5.fit(X_train, Y_train)
```

```python
#test the model for accuracy
predictions = model5.predict(X_test)
print("Model predicted", predictions)
print("Expected",Y_test)
```

```python
#Accuracy in percentage
from sklearn.metrics import accuracy_score
print("Accuracy", accuracy_score(Y_test, predictions))
```
