```python
#loading the dataset into a dataframe
import pandas
data = pandas.read_csv("datasets/bank.csv")
data

```

```python
#check whether the dataset contain empty slot
data.isnull().sum()
```

```python
#we notice that self_emplooyed column contains 32 empty slots.Since thhis column contains categorical data,we need to fill the empty slots with categorical value
data["Self_Employed"].fillna("unknown", inplace=True)
```

```python
data.isnull().sum()
```

```python
#showing the first 50 records of the new dataframe
data.head(50)
```

```python
data["LoanAmount"]
```

```python
#since the loan amount column contains quantitive data,we shall find the mean of all other entries and replace the empty slots with the mean.
mean_loan = data["LoanAmount"].mean()
mean_loan
```

```python
#use the mean to fill the null value
data["LoanAmount"].fillna(mean_loan, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#check the versions that exists inside of a matplot lib module
plt.style.available
```

```python
#specify one of the version
plt.style.use("seaborn-v0_8")
```

```python
#A count plot is a univariate bar graph.it means it only allows us to draw a bargraph by passing only one parameter
sns.countplot(x=data["Gender"],palette="autumn")
plt.title("Gender Distribution")
plt.xlabel("Different Gender Categories")
plt.ylabel("Distribution in Numbers")
```

```python
#creating a barplot.
#A barplot is multivariate i.e it accepts one to draw a bar graph using different columns
#A barplot is drawn by use of a categorical collumn against a quantitative column
import numpy

sns.barplot(x="Gender",y="LoanAmount", data=data, estimator=numpy.mean, palette="Greens", hue="Property_Area")
plt.title("Distribution of Gender against Loan Amount")
```

```python
#Swarmplot: It helps tto visualize where more members of a certain group lies.
sns.swarmplot(x="Gender", y="ApplicantIncome",data=data, palette="winter",hue="Loan_Status")
plt.title("Distribution of Gender against Applicant Income")
```

```python
#scatterplot: It is also used to show the distribution of different entities of a column
sns.scatterplot(x="LoanAmount", y="ApplicantIncome", data=data, palette="magma", hue="Education")
```
