```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8")
```

```python
sns.countplot(x=data["City"],palette="spring")
plt.title("Cities with our Customers")
plt.xlabel("Cities")
plt.ylabel("Number of Customers")
```

```python
sns.countplot(x=data["Customer type"],palette="summer")
plt.title("Cities with our Customers")
plt.xlabel("Branches")
plt.ylabel("Number of Customers")
```

```python
import numpy
import seaborn as sns


sns.barplot(x="City",y="Total", data=data, estimator=numpy.mean, palette="autumn", hue="Customer type")
plt.title("Total number of clients in different cities")
```

```python
import numpy
import seaborn as sns


sns.barplot(x="Payment",y="Tax 5%", data=data, estimator=numpy.mean, palette="winter", hue="Product line")
plt.title("Tax evaluation of members in regards various product lines.")
```

```python
sns.swarmplot(x="Gender", y="cogs",data=data, palette="spring",hue="Customer type")
plt.title("Distribution of Gender against cogs")
```

```python
sns.swarmplot(x="City", y="Rating",data=data, palette="autumn",hue="Branch")
plt.title("A swarmplot showing the rating in different cities")
```
